## Simple Majority Voting
With simple majority voting, you choose the most commonly predicted label amongst the classifiers as the ensemble classification.

### Lab Task 1
Consider the following example classifications of a test instance $X$ from an ensemble of 5 classifiers:

|Classifier |Prediction|
|-|-|
|1 |yes|
|2 |yes|
|3 |no|
|4 |yes|
|5 |no|

What is the ensemble classification of $X$ using simple majority voting?
* yes: 1 + 1 + 0 + 1 + 0 = 3 votes for yes
* no: 0 + 0 + 1 + 0 + 1 = 2 votes for no
* => yes wins!

## Weighted Majority Voting
With weighted majority voting, you use validation set accuracy results to weight votes.

### Lab Task 2
Continuing from the previous lab task, suppose this table stores the validation set accuracy results for each classifier:

|Classifier |Predictive Accuracy|
|-|-|
|1 |60%|
|2 |50%|
|3 |70%|
|4 |30%|
|5 |80%|

What is the ensemble classification of $X$ using weighted majority voting?
* yes: 0.6 + 0.5 + 0.0 + 0.3 + 0.0 = 1.4 votes for yes
* no: 0.0 + 0.0 + 0.7 + 0.0 + 0.8 = 1.5 votes for no
* => no wins!
    
There is a potential problem ... 

Q: Can you spot it?
* Since there are multiple labels (e.g., yes, no) each classifier may be good at predicting some but not others
* Can still lead to good overall accuracy (depending on distribution of labels)

## "Track Record" Voting
With track record voting, you use each classifier's track record of prediction to weight votes, giving each classifier a "split vote"

### Lab Task 4
Suppose the following table shows each classifier's validation set track records for predicting yes and no:

|classifier |prediction |yes track record|no track record|
|-|-|-|-|
|1 |yes |70% |30%|
|1 |no |50% |50%|
|2 |yes |60% |40%|
|2 |no |50% |50%|
|3 |yes |50% |50%|
|3 |no |20% |80%|
|4 |yes |80% |20%|
|4 |no |70% |30%|
|5 |yes |85% |15%|
|5 |no |33% |66%|

What is the ensemble classification of X using track record voting? Recall: 1 predicted yes, 2 predicted yes, 3 predicted no, 4 predicted yes, and 5 predicted no

|classifier |votes for yes |votes for no|
|-|-|-|
|1 |0.7 |0.3|
|2 |0.6 |0.4|
|3 |0.2 |0.8|
|4 |0.8 |0.2|
|5 |0.33 |0.66|
|total|2.63 |2.36|

=> yes wins again!
